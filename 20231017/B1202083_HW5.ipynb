```python
import random
ans=random.sample(range(0,10),4)
print(ans)
num=0
a=0
while(a!=4):
  a=b=n=0
  g=list(input("請輸入四個不重複之數字"))
  if len(g)>4:
    print("")
    print("輸入數字過多,")
  else:
    for i in g:
      if int(g[n])==ans[n]:
        a+=1
      else:
        if int(i) in ans:
          b+=1
      n+=1
    output=",".join(g).replace(",","")
    print(f'{output}:{a}A{b}B')
    print("")
    num+=1
print(f'猜對了，總共用了{num}次')
```
