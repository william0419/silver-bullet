```python
def BMI(height,weight):
    b=(weight/((height/100)**2))
    return b
try:
  height=float(input("輸入身高(公分)"))
  weight=float(input("輸入體重"))
  print(f'BMI:{BMI(height,weight)}')
  if(BMI(height,weight)<18.5):
    print('體位:過輕')
  elif(18.5<=BMI(height,weight)<24):
    print('體位:適中')
  elif(24<=BMI(height,weight)<27):
    print('體位:過重')
  else:
    print('體位:肥胖')
except:
  print("輸入內容錯誤")
```
