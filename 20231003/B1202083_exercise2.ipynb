```python
w=input ('輸入體重(kg):')
h=input ('輸入身高(m):')
w=float (w)
h=float (h)
w/=(h**2)
print('BMI:',w)
if w<18.5:
  print('體位:過輕')
elif 18.5<=w<24:
  print('體位:適中')
elif 24<=w<27:
  print('體位:過重')
else:
  print('體位:肥胖')
```
