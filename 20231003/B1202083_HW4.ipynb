```python
a=float(input('請輸入數字a '))
b=float(input('請輸入數字b '))
c=input('請輸入運算符號')
if c=='+':
  print(a+b)
elif c=='-':
  print(a-b)
elif c=='*':
  print(a*b)
else:
  print(a/b)
```

```python
y=int(input('請輸入年分'))
if y<=1582:
  if y%4==0:
    print('是閏年')
  else:
    print('是平年')
elif y%4==0 and y%100!=0:
  print('是閏年')
elif y%400==0:
  print('是閏年')
else:
  print('是平年')
```
